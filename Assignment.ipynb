```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/0311Chetan/Codes/main/pubg-weapon-stats.csv')
df
```

```python
# 1. Weapon name v/s Bullet type:
sns.barplot(x = df['Weapon Name'],y = df['Bullet Type'])
plt.xticks(rotation = 90)
```

```python
#2. Weapon Name v/s Damage :
sns.barplot(x = df['Weapon Name'],y = df['Damage'])
plt.xticks(rotation = 90)
```

```python
#3. Visualisation of Bullet Speed:
sns.distplot(df['Bullet Speed'])
```

```python
#4. Visualisation of Range:
sns.distplot(df['Range'])
```

```python
#5. Weapon Name v/s Magazine Capacity :
sns.barplot(x = df['Weapon Name'],y = df['Magazine Capacity'])
plt.xticks(rotation = 90)
```

```python
#6. Weapon Name v/s Weapon Type :
plt.scatter(df['Weapon Name'],df['Weapon Type'])
plt.xticks(rotation = 90)
```

```python
#7. Weapon Name v/s Shots to Kill (Chest) :
sns.barplot(x = df['Weapon Name'],y = df['Shots to Kill (Chest)'])
plt.xticks(rotation = 90)
```

```python
#8. Weapon Name v/s Shots to Kill (Head) :
sns.barplot(x = df['Weapon Name'],y = df['Shots to Kill (Head)'])
plt.xticks(rotation = 90)
```

```python
#10. Weapon name v/s Damage per Second :
sns.barplot(x = df['Weapon Name'],y = df['Damage Per Second'])
plt.xticks(rotation = 90)
```

```python
#11. Bullet speed v/s Damage :
sns.barplot(x= df['Bullet Speed'],y =df['Damage'])
plt.xticks(rotation = 90)
```
